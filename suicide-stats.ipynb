```python
# Quantidade de suic√≠dios no Brasil desde 1979.
#
# Dataset: https://www.kaggle.com/szamil/who-suicide-statistics
#
# https://github.com/pandas-dev/pandas/blob/master/doc/cheatsheet/Pandas_Cheat_Sheet.pdf
# https://s3.amazonaws.com/assets.datacamp.com/blog_assets/Python_Seaborn_Cheat_Sheet.pdf
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')

low_memory = False

sns.set(style="darkgrid")
```

```python
world_df = pd.read_csv("datasets/who_suicide_statistics.csv")

world_df.rename(columns={'country': 'pais',
                   'year': 'ano',
                   'sex': 'sexo',
                   'age': 'idade',
                   'suicides_no': 'suicidios',
                   'population': 'populacao'
                   }, inplace=True)

# traduz os valores de sexo
world_df = world_df.replace({
    "male": "homem",
    "female": "mulher"
})

# traduz os valores de faixa de idade
world_df = world_df.replace({
    "5-14 years": "5-14 anos",
    "15-24 years": "15-24 anos",
    "25-34 years": "25-34 anos",
    "35-54 years": "35-54 anos",
    "55-74 years": "55-74 anos",
    "75+ years": "+75 anos"
})

# filtro por Brasil
brazil_df = world_df.query('pais == "Brazil"')
```

```python
# m√©dia de suic√≠dios ao ano por sexo com intervalo de confian√ßa no brasil

f, ax = plt.subplots(figsize=(20,8))
sns.lineplot(ax=ax, x="ano", y="suicidios", hue="sexo", style="sexo", markers=True, dashes=False, data=brazil_df);
```

```python
# quantidade de suic√≠dios por faixa et√°ria agrupados por sexo no brasil

f, ax = plt.subplots(figsize=(20,8))
sns.swarmplot(ax=ax, x="idade", y="suicidios", hue="sexo", data=brazil_df);
```

```python
# quantidade de suic√≠dios por faixa et√°ria agrupados por sexo no brasil

f, ax = plt.subplots(figsize=(20,8))

g = sns.barplot(ax=ax, x="idade", y="suicidios", hue="sexo", data=brazil_df)
```

```python
# quantidade de suic√≠dios por faixa et√°ria agrupados por sexo no mundo

f, ax = plt.subplots(figsize=(20,8))

g = sns.barplot(ax=ax, x="idade", y="suicidios", hue="sexo", data=world_df)
```

```python
# quantidade de suic√≠dios por faixa et√°ria agrupados por sexo no mundo com intervalo de confian√ßa

f, ax = plt.subplots(figsize=(20,8))
# sns.barplot(ax=ax, x="idade", y="suicidios", hue="sexo", data=world_df);
sns.lineplot(ax=ax, x="idade", y="suicidios", hue="sexo", style="sexo", markers=True, dashes=False, data=brazil_df);
```

```python
# quantidade de suic√≠dios por faixa et√°ria agrupados por sexo no mundo

f, ax = plt.subplots(figsize=(20,8))
sns.swarmplot(ax=ax, x="idade", y="suicidios", hue="sexo", data=world_df);
```
