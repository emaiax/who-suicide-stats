```python
import pandas as pd
# import matplotlib.pyplot as plt
import seaborn as sns
# get_ipython().run_line_magic('matplotlib', 'inline')

low_memory = False

sns.set(style="darkgrid")
```

```python
df = pd.read_csv("datasets/who_suicide_statistics.csv")

df.rename(columns={'country': 'pais',
                   'year': 'ano',
                   'sex': 'sexo',
                   'age': 'idade',
                   'suicides_no': 'suicidios',
                   'population': 'populacao'
                   }, inplace=True)

# filtro por Brasil
# traduz os valores de sexo female/male para mulher/homem para facilitar a leitura do gráfico
brazil_ds = df.query('pais == "Brazil"').replace({"male": "homem", "female": "mulher"})
```

```python
# média de suicídios ao ano por sexo com intervalo de confiança

f, ax = plt.subplots(figsize=(15,8))
sns.lineplot(ax=ax, x="ano", y="suicidios", hue="sexo", style="sexo", markers=True, dashes=False, data=brazil_ds);
```

```python
# suicídios ao ano por faixa etária agrupados por sexo

f, ax = plt.subplots(figsize=(15,8))
sns.catplot(ax=ax, x="idade", y="suicidios", kind="swarm", hue="sexo", data=brazil_ds);
```
